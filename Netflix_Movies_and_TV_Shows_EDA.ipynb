```python
last_10_years_df = ntflx[["type","release_year","country","director","listed_in"]]
last_10_years_df=last_10_years_df[last_10_years_df["release_year"] >=2012]
last_10_years_df
```

```python
last_10_years_df.groupby("release_year")["type"].value_counts()
```

```python
plt.figure(figsize=(14,8))
sns.countplot(x="release_year", data=last_10_years_df, hue="type", palette="rocket")
```

```python
plt.figure(figsize=(14,8))
sns.countplot(x="release_year", data=last_10_years_df, palette="icefire")
```

```python
last_10_years_df_1 =last_10_years_df.groupby("release_year")["type"].size().reset_index()
last_10_years_df_1 = pd.DataFrame(last_10_years_df_1)
last_10_years_df_1
```

```python
last_10_years_df_1.rename(columns={"type" : "Total Movies/Shows"},inplace=True)
last_10_years_df_1
```

```python
plt.figure(figsize=(14, 8))
plot = sns.lineplot(x="release_year",y="Total Movies/Shows",data=last_10_years_df_1, linewidth=3,color='Red')
plot.set(xlabel='Release Year', ylabel='Total Movies/Shows', title='Trend of Content Produced over Ten Years on Netflix')
plt.show()
```

```python
top_10_countries = ntflx['country'].value_counts().head(10)
top_10_countries = pd.DataFrame(top_10_countries)
top_10_countries
```

```python
top_10_countries.rename(columns={"country" : "Total Content"},inplace=True)
top_10_countries
```

```python
plt.figure(figsize=(14, 8))
plot = sns.barplot(x=top_10_countries.index,y="Total Content",data=top_10_countries, linewidth=3,palette='dark:salmon_r')
plot.set(xlabel='Countries', ylabel='Total Movies/Shows', title='Top 10 countries with highest number of Movies/TV Shows on Netflix')
plt.show()
```

```python
ntflx_movie = ntflx[ntflx['type']=='Movie']
ntflx_movie
```

```python
ntflx_movie.duration = ntflx_movie.duration.apply(lambda x :int(x.split()[0]))
```

```python
ntflx_movie.head()
```

```python
ntflx_movie_10=ntflx_movie.sort_values('duration', ascending=False).head(10)
ntflx_movie_10
```

```python
plt.figure(figsize=(14, 8))
plot=sns.barplot(x='title',y="duration",data=ntflx_movie_10, linewidth=3,palette='viridis')
plot.set(xlabel='Title', ylabel='Duration', title='Top 10 Movies with highest Duration in mins')
plt.xticks(rotation=45)
plt.show()
```

```python
ntflx_tv = ntflx[ntflx['type']=='TV Show']
ntflx_tv
```

```python
ntflx_tv.duration = ntflx_tv.duration.apply(lambda x :int(x.split()[0]))
```

```python
ntflx_tv.head()
```

```python
ntflx_tv_10=ntflx_tv.sort_values('duration', ascending=False).head(10)
ntflx_tv_10
```

```python
plt.figure(figsize=(14, 8))
plot=sns.barplot(x='title',y="duration",data=ntflx_tv_10, linewidth=3,palette='magma')
plot.set(xlabel='Title', ylabel='Number of Seasons', title='Top 10 TV Shows with highest number of Seasons')
plt.xticks(rotation=45)
plt.show()
```
